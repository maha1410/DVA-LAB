### Grouping Pivot table

```python
age = pd.cut(df['age'], [0, 18, 80])
df.pivot_table('survived', ['sex', age], 'class')
```

We can apply this same strategy when working with the columns as well; let’s add info on the fare paid using
pd.qcut to automatically compute quantiles

```python
fare = pd.qcut(df['fare'], 2)
df.pivot_table('survived', ['sex', age], [fare, 'class'])
```

The result is a four-dimensional aggregation with hierarchical indices

### Multiple Aggregate Functions

```python
df.pivot_table(index='sex', columns='class',
aggfunc={'survived':sum, 'fare':'mean'})
```

### Melt - from Pivot Table to long or tidy format

The melt function performs the inverse operation of a pivot . This can be used to make your frame longer,
i.e. to make a tidy version of your data.

```python
pivoted = df.pivot_table(index='sex', columns='pclass', values='fare').reset_index()
pivoted.columns.name = None
```

```python
pivoted
```

Assume we have a DataFrame like the above. The observations (the average Fare people payed) are spread
over different columns. In a tidy dataset, each observation is stored in one row. To obtain this, we can use the 
melt function:

```python
pd.melt(pivoted)
```

As you can see above, the melt function puts all column labels in one column, and all values in a second
column.

In this case, this is not fully what we want. We would like to keep the 'Sex' column separately:

```python
pd.melt(pivoted, id_vars=['sex']) #, var_name='pclass', value_name='fare')
```

### Reshaping with stack and unstack

Before we speak about hierarchical index , first check it in practice on the following dummy example:

```python
df2 = pd.DataFrame({'A':['one', 'one', 'two', 'two'], 
 'B':['a', 'b', 'a', 'b'], 
 'C':range(4)})
df2
```

To use stack / unstack , we need the values we want to shift from rows to columns or the other way around as
the index:

```python
df2 = df2.set_index(['A', 'B']) # Indeed, you can combine two indices
df2
```

```python
result = df2['C'].unstack()
result
```

```python
df2 = result.stack().reset_index(name='C')
df2
```
