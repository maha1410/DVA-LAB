### Merging Historical Labor Data
It’s nice being able to see how the labor hours have shifted since 2000, but in order to see real trends emerge,
we want to be able to see as much historical data as possible. The data collection team was kind enough to send
data from 1950 to 2000, let’s load it in and take a look.

```python
historical = pd.read_csv('historical.csv', index_col=0)
historical.head()
```

```python
print("World rows & columns: ", world.shape)
print("Historical rows & columns: ", historical.shape)
```

### Merge historical dataframe with world dataframe and store in a new variable, world_historical

```python
world_historical = pd.merge(historical, world, left_index=True, right_index=True, how='right')
```

### Print size of world_historical dataframe

```python
print(world_historical.shape)
```

### Print top-5 of world_historical dataframe

```python
world_historical.head()
```

### Joining Historical Data
Now that we’ve done it the hard way and understand table merging conceptually, let’s try a more elegant
technique. Pandas has a clean method to join on indexes which is perfect for our situation.
### Use join method to join historical dataframe and world dataframe and store result in world_historical dataframe

```python
world_historical = historical.join(world, how='right')
world_historical.head()      # Print head of world_historical dataframe
```

### Plot, transposed world_historical dataframe

```python
world_historical.sort_index(inplace=True)
world_historical.transpose().plot(figsize=(15,10), colormap='rainbow', linewidth=2, title='Average Labor Hours Per Year')
plt.legend(loc='right', bbox_to_anchor=(1.15, 0.5))
plt.show()
```

### Which country worked longer hours per year?

```python
work=world.mean(axis=1)

long=max(world.mean(axis=1))

short=min(world.mean(axis=1))
```

```python
print("country worked longer hours per year : ",work[work == long].index[0])
```

### Which country worked shorter hours per year?

```python
print("country worked longer hours per year : ",work[work==short].index[0])
```
